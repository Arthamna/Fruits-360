```python
import os
import numpy as np
import gradio as gr
import tensorflow as tf
from tensorflow import keras
```

```python
IMG_SIZE = (100, 100)

TEST_PATH = "/media/blue/data/Projects/Fruits-360-local/fruits-360_100x100/fruits-360/Test"

MODEL_PATHS = {
    "CustomCNN": "Fruits360-CustomCNN.keras",
    "MobileNetV2": "Fruits360-MobileNetV2.keras",
}
```

```python
class_names = sorted(os.listdir(TEST_PATH))
print(f"Jumlah kelas: {len(class_names)}")

models = {}
for name, path in MODEL_PATHS.items():
    try:
        models[name] = keras.models.load_model(path)
        print(f"Model '{name}' berhasil dimuat dari {path}")
    except Exception as e:
        print(f"Gagal memuat model '{name}' dari {path}: {e}")


def predict(image, model_choice):
    if image is None:
        return "Silakan unggah gambar terlebih dahulu."

    model = models.get(model_choice)
    if model is None:
        return f"Model '{model_choice}' tidak tersedia. Cek path file model."

    img = tf.image.resize(image, IMG_SIZE)
    arr = tf.expand_dims(img, axis=0)

    preds = model.predict(arr, verbose=0)[0]

    top_idx = np.argsort(preds)[::-1][:5]
    return {class_names[i]: float(preds[i]) for i in top_idx}
```

```python
demo = gr.Interface(
    fn=predict,
    inputs=[
        gr.Image(type="numpy", label="Unggah gambar buah"),
        gr.Radio(
            choices=list(MODEL_PATHS.keys()),
            value="MobileNetV2",
            label="Pilih model",
        ),
    ],
    outputs=gr.Label(num_top_classes=5, label="Hasil prediksi (5 teratas)"),
    title="Klasifikasi Buah — Fruits-360",
    description="Unggah gambar buah, pilih model (CustomCNN atau MobileNetV2), lalu lihat prediksinya.",
)

if __name__ == "__main__":
    demo.launch()
```
